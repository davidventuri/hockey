```python
def rename_team(row):
    if row['Team'] in team_abbrev.keys():
        abbrev = team_abbrev[row['Team']]
        return abbrev
    else:
        return row['Team']
```

```python
# Change team name to abbreviation
toi_1516['Team'] = toi_1516.apply(rename_team, axis=1)
toi_1415['Team'] = toi_1415.apply(rename_team, axis=1)
toi_1314['Team'] = toi_1314.apply(rename_team, axis=1)
toi_1213['Team'] = toi_1213.apply(rename_team, axis=1)
toi_1112['Team'] = toi_1112.apply(rename_team, axis=1)
toi_1011['Team'] = toi_1011.apply(rename_team, axis=1)
toi_0910['Team'] = toi_0910.apply(rename_team, axis=1)
toi_0809['Team'] = toi_0809.apply(rename_team, axis=1)
toi_0708['Team'] = toi_0708.apply(rename_team, axis=1)
```

```python
def toi_string_to_int(row):
    toi = row["TOI"].split(":")
    minutes = float(toi[0])
    seconds_decimal = float(toi[1]) / 60.0
    return minutes + seconds_decimal
```

```python
# Convert TOI to integer to later calculate xG60
toi_1516['TOI'] = toi_1516.apply(toi_string_to_int, axis=1)
toi_1415['TOI'] = toi_1415.apply(toi_string_to_int, axis=1)
toi_1314['TOI'] = toi_1314.apply(toi_string_to_int, axis=1)
toi_1213['TOI'] = toi_1213.apply(toi_string_to_int, axis=1)
toi_1112['TOI'] = toi_1112.apply(toi_string_to_int, axis=1)
toi_1011['TOI'] = toi_1011.apply(toi_string_to_int, axis=1)
toi_0910['TOI'] = toi_0910.apply(toi_string_to_int, axis=1)
toi_0809['TOI'] = toi_0809.apply(toi_string_to_int, axis=1)
toi_0708['TOI'] = toi_0708.apply(toi_string_to_int, axis=1)
```
